```python
df = pd.read_csv(
    './submission_data.csv',
    low_memory=False
)
```

```python
# Assuming 'df' is your DataFrame

# Step 1: Combine 'cluster_nl' and 'launch_date' into a single column
df['combined'] = df['cluster_nl'].astype(str) + "_" + df['launch_date'].astype(str)

# Step 2: Get unique combinations of 'cluster_nl' and 'launch_date'
unique_combinations = df[['combined']].drop_duplicates()

# Step 3: Check for duplicates in the 'combined' column
duplicates = df[df.duplicated(subset=['combined'], keep=False)]

# Print results
print("Unique Combinations:")
print(unique_combinations)

print("\nDuplicates:")
print(duplicates[['cluster_nl', 'launch_date', 'combined']])
```

```python
# Assuming 'df' is your DataFrame

# Step 1: Sort the DataFrame by 'date' to ensure the correct order
df_sorted = df[['date', 'launch_date', 'cluster_nl']].sort_values(by='date')

# Step 2: Group by 'cluster_nl' and show 'date' and 'launch_date' for each group
grouped_df = df_sorted.groupby('cluster_nl').apply(lambda x: x[['cluster_nl','date', 'launch_date']])

# Step 3: Reset the index to get a clean DataFrame
grouped_df = grouped_df.reset_index(drop=True)

# Print the result
print(grouped_df)
grouped_df.to_csv('cluster_launch_date.csv')
```
